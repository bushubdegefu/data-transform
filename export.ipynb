# Database Export Examples

This notebook demonstrates how to use `db_exporter.py` to export database tables to SQL or CSV files.


```python
from db import get_session, engine, Base
from db_exporter import export_table, export_table_to_csv, export_table_to_sql

# Ensure tables exist
Base.metadata.create_all(bind=engine)
print("Database tables ready.")
```

## Export to CSV

Export a table to CSV format.

```python
with get_session() as session:
    export_table_to_csv(
        session=session,
        table_name="customers",
        output_path="current-data/customers_export.csv"
    )
```

```python
with get_session() as session:
    export_table_to_csv(
        session=session,
        table_name="carrier",
        output_path="current-data/carrier_export.csv"
    )
```

## Export to SQL with Batched INSERT Statements

Export a table to SQL format with batched INSERT statements (default: 1000 rows per statement).

```python
with get_session() as session:
    export_table_to_sql(
        session=session,
        table_name="depot",
        output_path="current-data/depot_export.sql",
        batch_size=1000  # Group up to 1000 rows in a single INSERT statement
    )
```

## Export Specific Columns

You can also export only specific columns.

```python
with get_session() as session:
    export_table(
        session=session,
        table_name="customers",
        output_path="example_data/customers_partial.csv",
        format="csv",
        columns=["id", "customer_name", "email"]
    )
```

## Custom Batch Size for SQL Export

You can adjust the batch size for SQL exports.

```python
with get_session() as session:
    export_table(
        session=session,
        table_name="customers",
        output_path="example_data/customers_small_batch.sql",
        format="sql",
        batch_size=100  # Smaller batches
    )
```
