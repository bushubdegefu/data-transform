# Run SQL Inserts

This notebook reads SQL files containing `INSERT` statements and executes them against the database using the project's `db` connection.


```python
from pathlib import Path
from sqlalchemy import text
from db import get_session, engine, Base

# Ensure tables exist (optional, but good practice)
Base.metadata.create_all(bind=engine)
print("Database tables checked/created.")
```

```python
def execute_sql_file(file_path):
    path = Path(file_path)
    if not path.exists():
        print(f"File not found: {path}")
        return

    print(f"Reading SQL from {path}...")
    sql_content = path.read_text(encoding="utf-8")
    
    # Split into individual statements
    # This is a simple split by semicolon. For more complex SQL (e.g. with triggers/procedures), 
    # a more robust parser might be needed, but for simple INSERTs this is usually sufficient.
    statements = [s.strip() for s in sql_content.split(';') if s.strip()]
    
    print(f"Found {len(statements)} statements to execute.")
    
    with get_session() as session:
        for statement in statements:
            session.execute(text(statement))
        print("All statements executed successfully.")
```

```python
# Define the path to your SQL file
sql_file_path = "example_data/customers_example_output.sql"

# Execute
execute_sql_file(sql_file_path)
```
