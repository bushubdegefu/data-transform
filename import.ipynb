# Database Import Examples

This notebook demonstrates how to use `db_importer.py` to import SQL files into the database.

The import function reads SQL files (like those exported by `db_exporter.py`) and executes the INSERT statements.


```python
from db import get_session, engine, Base
from db_importer import import_sql, import_sql_file

# Ensure tables exist
Base.metadata.create_all(bind=engine)
print("Database tables ready.")
```

## Simple Import - Using Table Name

Import from the default location (`exported/{table_name}.sql`).

```python
# Import from exported/customers.sql
count = import_sql(table_name="customers")
print(f"Imported {count} statements")
```

## Import from Custom Path

Import from a specific SQL file path.

```python
# Import from a custom path
count = import_sql(sql_file_path="example_data/customers_example_output.sql")
print(f"Imported {count} statements")
```

## Manual Import with Session Control

For more control, you can use `import_sql_file()` with your own session.

```python
with get_session() as session:
    count = import_sql_file(
        session=session,
        sql_file_path="exported/customers.sql"
    )
    print(f"Imported {count} statements")
```

## Complete Export/Import Workflow

Example of exporting and then importing data.

```python
from db_exporter import export

# Export a table
export("customers", batch_size=1000)

# Later, import it back (e.g., to another database or after clearing the table)
count = import_sql(table_name="customers")
print(f"Exported and re-imported {count} statements")
```
